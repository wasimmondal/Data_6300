<h3 style="color:black;"> 6. Multiple Regression for ‘Sale Price</h3>

<h4 style="color:black;"> 6a: Creating the X and y datasets</h4>

```python
df_realestate.info()
```

```python
X = df_realestate[['Lot Frontage', 'Lot Area', 'Overall Quality', 'Overall Cond', 'Masonry/Veneer Area', 'Basement Finished Area', 'Basement Unfinished Area', '1st Floor Area', '2nd Floor Area', 'Bedrooms Above Grade', 'Fireplaces', 'Garage Cars', 'Porch Area', 'Full Bathrooms', 'Half Bathrooms', 'Extra Kitchens', 'Bathroom Percentage', 'Baths per Bedroom' ]]
```

```python
X.head(5)
```

```python
y = df_realestate['Sale Price']
```

```python
y.head(10)
```

<h4 style="color:black;"> 6b: Create a Multiple Regression Model</h4>

```python
X = sm.add_constant(X) # adding a constant
reg = sm.OLS(y, X).fit()
pred = reg.predict(X) 
resid = reg.resid
reg.summary()
```

<font color='blue'> Explain the results</font> 

Is the model significant? Yes, the model is significant because R-squared value is 0.857 which is reasonable. In ideal case, we expect R-squared to be close to 1.  

Are all of the features significant? If the p value is small, that means every features contributed to the model. As we can observe all the feature's p value is almost 0.00 except two features which are Baths per Bedroom and Fireplaces. It indicates that Baths per Bedroom and Fireplaces are not significantly contributing to the model. 

Is the model a strong relationship? Overall, for most of the features, model is a strong relationship.

<font color='blue'> Create the Varience Inflation Factor (VIF)</font> 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X.info()
```

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["Features"] = X.columns
round(vif, 1).sort_values(by = 'VIF Factor', ascending = False)
```

```python
#As we see from above, we note that ‘Bathroom Percentage’ has the highest VIF with Full Bathrooms. 
```

Why ‘Bathroom Percentage’ has the highest VIF ? VIF factor measures the effect of multicollinearity. Multicollinearity means two or more factors are higly correlated with each other. If the multicollinearity is present in a model, information can be redundant. It will cause the estimate of a coeficient to change depending on which other variables are included in the model. The p value the test for whather the coeficient is zero can change depedning on which other variables are present and error in prediction can be higher than necessary.
Bathroom Percentage has already information of Full Bathrooms. Hence, it has highest VIF factor adding multicollinearity.

<h3 style="color:black;"> 7. Summary and Explanation of All Steps</h3>

When we build a model, we should review all possible features that will be available at the time the model is needed. We can investigate  the multi-collinearity in the model and if we find any highly correlated or non-significant features, we should remove it.  

Now, question is how can we invetsigate collinearity among predictor variables? This we can easily do by calculating VIF. As we know, VIF is a measure of collinearity among predictor variables with a multiple regressions. It is calculated by taking the ratio of varianece of all a given model's betas divided by the variance of single beta if it were fit alone. After calculating VIF, if we find VIF factor for any feature is greater than 5, that feature may have some problem, and we can remove them. 

But removing features based on VIF factor may not totally solve the problem in some situations. As we know, we evaluate model's performance based on the R-squared. R-squared measures the amount of variations in the target variable that is explained by the model. The adjusted R-squared value adjusts this model by a factor related to the number of variables in the model. Besides R-squared, we can also look into P value and t value. If the p value is small, that means every features contributed to the model. 

In some cases, after removing high VIF factor feature, we find model's R-squared value does not change much. But, removing that feature with high VIF factor, remove any multicollinearity . 

Computing correlation matrix or heat map also helpful for deciding what features are inmportant to build a good model. If we find high correlation value in the correlation matrix or in the heat map, that means information is already there in one feature among them, We can remove other. 

Based on R-squared, we find Model 1 is more significant than model 2.
